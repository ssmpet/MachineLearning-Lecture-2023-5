```python
X_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv')
X_test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv')

y_train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv')
```

```python
drop_col = ['CustomerId', 'Surname']
X_train_drop = X_train.drop(columns=drop_col)
X_test_drop = X_test.drop(columns=drop_col)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_test_drop
```

```python
X_test_dummies = pd.get_dummies(X_test_drop)
X_test_dummies
```

```python
X_train_dummies = pd.get_dummies(X_train_drop)
X_train_dummies
```

```python
X_test_dummies = X_test_dummies[X_train_dummies.columns]
X_test_dummies
```

```python
help(train_test_split)
```

```python
y = y_train.Exited
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_train_dummies, y, test_size=0.33, random_state=42
)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
help(RandomForestClassifier)
```

```python
rf = RandomForestClassifier(random_state=23)
rf.fit(X_train, y_train)
```

```python
pred_train = rf.predict(X_train)
pred_train_prob = rf.predict_proba(X_train)[:, 1]
pred_train_prob
```

```python
pred_test = rf.predict(X_test)
pred_test_prob = rf.predict_proba(X_test)[:, 1]
pred_test_prob
```

```python
import sklearn.metrics
dir(sklearn.metrics)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, roc_auc_score
```

```python
accuracy_score(y_train, pred_train), accuracy_score(y_test, pred_test)
```

```python
f1_score(y_train, pred_train), f1_score(y_test, pred_test)
```

```python
recall_score(y_train, pred_train), recall_score(y_test, pred_test)
```

```python
precision_score(y_train, pred_train), precision_score(y_test, pred_test)
```

```python
roc_auc_score(y_train, pred_train), roc_auc_score(y_test, pred_test)
```
