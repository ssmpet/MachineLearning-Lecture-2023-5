```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
filename = 'web/static/data/melon_song_v3.csv'
plist_filename1 = 'web/static/data/melon_playlist1.csv'
plist_filename2 = 'web/static/data/melon_playlist2_v2.csv'
```

```python
df = pd.read_csv(filename)
plist1 = pd.read_csv(plist_filename1)
plist2 = pd.read_csv(plist_filename2)
df.shape, plist1.shape, plist2.shape
```

```python
df.songId = df.songId.astype(str)
plist1.plylstSeq = plist1.plylstSeq.astype(str)
plist2.songId = plist2.songId.astype(str)

df.genre.fillna('-', inplace=True)
df.fillna('', inplace=True)
df.lyric = df.lyric.str.replace('\r', '')
df.head(1)
```

```python
# songId 별로 playlist 에 들어있는 개수
df['p_count'] = df.songId.apply(lambda x : len(plist1[(plist1.songIds + ' ').str.contains(str(x)+ ' ')]))
```

```python
# playlist에 가장 많이 들어가 있는 곡은? 
df.sort_values('p_count', ascending=False)[['songId', 'title', 'artist', 'album', 'date', 'p_count', 'img']].head(20)
```

```python
# 저장하기
df.to_csv('web/static/data/best_song2.csv', index=False)
```
