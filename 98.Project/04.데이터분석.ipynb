```python
import numpy as np
import pandas as pd
```

```python
filename = 'web/static/data/melon_song_v3.csv'
plist_filename1 = 'data/melon_playlist1.csv'
plist_filename2 = 'data/melon_playlist2_v2.csv'
```

```python
df = pd.read_csv(filename)
plist1 = pd.read_csv(plist_filename1)
plist2 = pd.read_csv(plist_filename2)
df.shape, plist1.shape, plist2.shape
```

```python
# df.date.fillna(0, inplace=True)
# df.date = df.date.astype(int).astype(str)
df.songId = df.songId.astype(str)
plist1.plylstSeq = plist1.plylstSeq.astype(str)
plist2.songId = plist2.songId.astype(str)

df.fillna('', inplace=True)
df.lyric = df.lyric.str.replace('\r', '')
df.head(1)
```

```python
df.isna().sum()
```

```python
for val in df.genre.unique():
    print(val)
```

```python
# 장르별 노래 총합
df.pivot_table('songId', 'genre', aggfunc='count').sort_values('songId', ascending=False).head(10)
```

```python
# 발매년도별 노래 총합
df['year'] = df.date.str[:4]
```

```python

df.pivot_table('songId', 'year', aggfunc='count').sort_values('songId', ascending=False).head(10)
```

```python
df.columns
```

```python
plist1.songIds
```

```python
# songId 별로 playlist 에 들어있는 개수
df['p_count'] = df.songId.apply(lambda x : len(plist1[plist1.songIds.str.contains(str(x))]))
```

```python
# playlist에 가장 많이 들어가 있는 곡은? 
df.sort_values('p_count', ascending=False)[['songId', 'title', 'artist', 'genre', 'comment', 'like', 'year', 'p_count']].head(20)
```

```python
# 좋아요가 가장 많이 들어가 있는 곡은? 
df.sort_values('like', ascending=False)[['songId', 'title', 'artist', 'genre', 'comment', 'like', 'year', 'p_count']].head(20)
# df.sort_values('like')
```

```python
# 댓글 수가 가장 많이 들어가 있는 곡은? 
df.sort_values('comment', ascending=False)[['songId', 'title', 'artist', 'genre', 'comment', 'like', 'year', 'p_count']].head(20)
```

```python
# 년도별 댓글 수가 가장 많은 곡
result = df.groupby('year')['comment'].idxmax()
max_rows = df.loc[result]
max_rows[['songId', 'title', 'artist', 'comment', 'like', 'year']].sort_values('year', ascending=False)
```

```python
# 년도별 좋아요 수가 가장 많은 곡
result = df.groupby('year')['like'].idxmax()
max_rows = df.loc[result]
max_rows[['songId', 'title', 'artist', 'comment', 'like', 'year']].sort_values('year', ascending=False)
```

```python
# 년도별 곡들 중 가장 많은 장르
res = df.groupby('year')['genre'].apply(lambda x: x.value_counts().idxmax())
df_genre = pd.DataFrame(res)
df_genre.sort_index(ascending=False, inplace=True)
df_genre.reset_index(inplace=True)
df_genre.columns = ['년도', '장르']
```

```python
df_genre
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(12, 10))
plt.barh(df_genre['년도'], df_genre['장르'])
plt.xticks(rotation=45)
plt.xlabel('년도')
plt.ylabel('장르')
plt.title('년도별 가장 많은 장르')
plt.show()
```
