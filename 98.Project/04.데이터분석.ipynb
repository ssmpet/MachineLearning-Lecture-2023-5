```python
import numpy as np
import pandas as pd
```

```python
filename1 = 'data/melon_genre_ost_230527.csv'
filename2 = 'data/melon_year_concat.csv'
filename3 = 'data/melon_genre.csv'
plist_filename1 = 'data/melon_playlist1.csv'
plist_filename2 = 'data/melon_playlist2.csv'
```

```python
df1 = pd.read_csv(filename1)
df2 = pd.read_csv(filename2)
df3 = pd.read_csv(filename3)
plist1 = pd.read_csv(plist_filename1)
plist2 = pd.read_csv(plist_filename2)
df1.shape, df2.shape, df3.shape, plist1.shape, plist2.shape
```

```python
df = pd.concat([df1, df2, df3])
df.drop_duplicates(subset=['songId'], inplace=True)
df.date.fillna(0, inplace=True)
df.date = df.date.astype(int).astype(str)
df.shape, df.songId.nunique()
```

```python
df.isna().sum()
```

```python
for val in df.genre.unique():
    print(val)
```

```python
# 장르별 노래 총합
df.pivot_table('songId', 'genre', aggfunc='count').sort_values('songId', ascending=False).head(10)
```

```python
# 발매년도별 노래 총합
df['year'] = df.date.str[:4]
```

```python

df.pivot_table('songId', 'year', aggfunc='count').sort_values('songId', ascending=False).head(10)
```

```python
# songId 별로 playlist 에 들어있는 개수
df['p_count'] = df.songId.apply(lambda x : len(plist1[plist1.songIds.str.contains(str(x))]))
```

```python
# playlist에 가장 많이 들어가 있는 곡은? 
df.sort_values('p_count', ascending=False)[['songId', 'title', 'artist', 'genre', 'comment', 'like', 'year', 'p_count']].head(20)
```
